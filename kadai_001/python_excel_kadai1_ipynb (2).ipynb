```python
import openpyxl
#新規ワークブックを作成
workbook = openpyxl.Workbook()
#アクティブなシートを取得する
ws = workbook.active

#セルに値を書き込む
ws.cell(row=2, column=1, value='請求書')
ws.cell(row=4, column=1, value='株式会社ABC')
ws.cell(row=5, column=1, value='〒101-0022 東京都千代田区神田練塀町300')
ws.cell(row=6, column=1, value='TEL:03-1234-5678 FAX:03-1234-5678')
ws.cell(row=7, column=1, value='担当者名:鈴木一郎 様')

ws.cell(row=4, column=5, value='No.')
ws.cell(row=5, column=5, value='日付')
ws.cell(row=4, column=6, value='0001')
ws.cell(row=5, column=6, value='2023/06/06')


#表のヘッダー行を書き込む
header = ['商品名', '単価', '数量', '売上']
ws.append(header)

#表のデータを書き込む
data = [

    ['商品A',2, 10000, 20000],
    ['商品B', 1, 15000, 15000]
    ]
#for文でdataの各行をワークシートに追加する
for row in data:
    ws.append(row)

#それぞれの金額を変数「total」「fee」「tatal_price」に代入する
total = 35000
fee = 35000 * 0.1
total_price = total + fee
#合計金額のデータを'D11'に書き込む
ws['D11'] = total
#表のデータを書き込む
data2 = [
    ['合計',"", "", total,],
    ['消費税',"", "", fee],
    ['合計金額', "", "", total_price]
    ]

#for文でdataを各行のワークシートに追加する
for row in data2:
    ws.append(row)

#行を追加する
ws.insert_rows(8, 2)
ws.insert_rows(14, 1)
ws.insert_cols(1, 1)

#セルを結合する（'B5:E5'と'B6:E6')
ws.merge_cells('B5:E5')
ws.merge_cells('B6:E6')

#datetimeモジュールをインポート
import datetime

#指定した日時を取得する
date = datetime.datetime.strptime("2023-05-20", "%Y-%m-%d")

#日付のtodayオブジェクトを、特定のフォーマットで文字列にする
backup_filename = f"backup{date.strftime('%Y%m%d')}.xlsx"

workbook.save(backup_filename)







```
